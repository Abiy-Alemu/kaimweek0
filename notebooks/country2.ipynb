```python
# Function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers in key columns
outliers_ghi = detect_outliers_iqr(df, 'GHI')
outliers_dni = detect_outliers_iqr(df, 'DNI')
outliers_dhi = detect_outliers_iqr(df, 'DHI')
outliers_moda = detect_outliers_iqr(df, 'ModA')
outliers_modb = detect_outliers_iqr(df, 'ModB')
outliers_ws = detect_outliers_iqr(df, 'WS')
outliers_wsgust = detect_outliers_iqr(df, 'WSgust')

print("\nOutliers in GHI:\n", outliers_ghi)
print("\nOutliers in DNI:\n", outliers_dni)
print("\nOutliers in DHI:\n", outliers_dhi)
print("\nOutliers in ModA:\n", outliers_moda)
print("\nOutliers in ModB:\n", outliers_modb)
print("\nOutliers in WS:\n", outliers_ws)
print("\nOutliers in WSgust:\n", outliers_wsgust)
```

```python
from scipy.stats import zscore

# Select relevant columns for Z-score calculation
columns_to_analyze = ['GHI', 'DNI', 'DHI', 'WS', 'Tamb', 'TModA', 'TModB']

# Calculate Z-scores
df_z_scores = df[columns_to_analyze].apply(zscore)

# Display the Z-scores
print(df_z_scores.head())
```

```python
# Define threshold for flagging outliers
threshold = 3

# Identify outliers
outliers = (df_z_scores.abs() > threshold)

# Display the outliers
print(outliers.head())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('sierraleone-bumbuna.csv', parse_dates=['Timestamp'])

# Check for missing values
missing_values = df.isnull().sum()

# Display columns with missing values
print("Missing Values:\n", missing_values[missing_values > 0])

# Check for anomalies in specific columns
# For example, checking for negative values in columns where they shouldn't exist
anomalies = {
    'GHI': df['GHI'] < 0,
    'DNI': df['DNI'] < 0,
    'DHI': df['DHI'] < 0,
    'WS': df['WS'] < 0,
    'Tamb': df['Tamb'] < -50,  # Assuming temperatures below -50Â°C are anomalies
    'TModA': df['TModA'] < -50,
    'TModB': df['TModB'] < -50
}

# Display anomalies
for column, condition in anomalies.items():
    if condition.any():
        print(f"Anomalies in {column}:\n", df[condition][column])
```

```python
# Drop columns with all null values
df = df.dropna(axis=1, how='all')

# For columns with some missing values, impute or drop rows
# Example: Impute missing values in temperature columns with mean
temperature_columns = ['Tamb', 'TModA', 'TModB']
for column in temperature_columns:
    df[column].fillna(df[column].mean(), inplace=True)

# Example: Drop rows with missing values in critical columns
df.dropna(subset=['GHI', 'DNI', 'DHI'], inplace=True)

# Check if there are any remaining missing values
print("Remaining Missing Values:\n", df.isnull().sum())
```

```python
# Filter out anomalies for specific columns
df = df[~anomalies['GHI']]
df = df[~anomalies['DNI']]
df = df[~anomalies['DHI']]
df = df[~anomalies['WS']]
df = df[df['Tamb'] >= -50]
df = df[df['TModA'] >= -50]
df = df[df['TModB'] >= -50]

# Verify that anomalies have been handled
print("Data Shape After Removing Anomalies:", df.shape)
```

```python
# Verify data summary
print(df.describe())

# Save cleaned data to a new CSV file
df.to_csv('sierraleone-bumbuna-cleaned.csv', index=False)
```

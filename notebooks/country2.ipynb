```python
# Function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers in key columns
outliers_ghi = detect_outliers_iqr(df, 'GHI')
outliers_dni = detect_outliers_iqr(df, 'DNI')
outliers_dhi = detect_outliers_iqr(df, 'DHI')
outliers_moda = detect_outliers_iqr(df, 'ModA')
outliers_modb = detect_outliers_iqr(df, 'ModB')
outliers_ws = detect_outliers_iqr(df, 'WS')
outliers_wsgust = detect_outliers_iqr(df, 'WSgust')

print("\nOutliers in GHI:\n", outliers_ghi)
print("\nOutliers in DNI:\n", outliers_dni)
print("\nOutliers in DHI:\n", outliers_dhi)
print("\nOutliers in ModA:\n", outliers_moda)
print("\nOutliers in ModB:\n", outliers_modb)
print("\nOutliers in WS:\n", outliers_ws)
print("\nOutliers in WSgust:\n", outliers_wsgust)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset and parse dates
df = pd.read_csv('sierraleone-bumbuna.csv', parse_dates=['Timestamp'])

# Set the Timestamp column as the index for easy time-based indexing
df.set_index('Timestamp', inplace=True)

# Display the first few rows to verify
print(df.head())
```

```python
plt.figure(figsize=(15, 10))

# Plot GHI
plt.subplot(2, 2, 1)
plt.plot(df.index, df['GHI'], color='orange')
plt.title('GHI Over Time')
plt.xlabel('Time')
plt.ylabel('GHI (W/m²)')

# Plot DNI
plt.subplot(2, 2, 2)
plt.plot(df.index, df['DNI'], color='blue')
plt.title('DNI Over Time')
plt.xlabel('Time')
plt.ylabel('DNI (W/m²)')

# Plot DHI
plt.subplot(2, 2, 3)
plt.plot(df.index, df['DHI'], color='green')
plt.title('DHI Over Time')
plt.xlabel('Time')
plt.ylabel('DHI (W/m²)')

# Plot Tamb (ambient temperature)
plt.subplot(2, 2, 4)
plt.plot(df.index, df['Tamb'], color='red')
plt.title('Tamb Over Time')
plt.xlabel('Time')
plt.ylabel('Tamb (°C)')

plt.tight_layout()
plt.show()
```

```python
# Plot ModA and ModB with respect to the Cleaning column
plt.figure(figsize=(15, 5))

# Plot ModA
plt.plot(df.index, df['ModA'], label='ModA', color='blue')
plt.plot(df.index[df['Cleaning'] == 1], df['ModA'][df['Cleaning'] == 1], 'ro', markersize=5, label='Cleaning Event')

# Plot ModB
plt.plot(df.index, df['ModB'], label='ModB', color='green')
plt.plot(df.index[df['Cleaning'] == 1], df['ModB'][df['Cleaning'] == 1], 'ro', markersize=5)

plt.title('Sensor Readings (ModA, ModB) Over Time with Cleaning Events')
plt.xlabel('Time')
plt.ylabel('Sensor Reading')
plt.legend()
plt.show()
```

```python
# Group by month and plot monthly trends for GHI, DNI, DHI, and Tamb
monthly_data = df.resample('M').mean()

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.plot(monthly_data.index, monthly_data['GHI'], color='orange')
plt.title('Monthly Average GHI')
plt.xlabel('Month')
plt.ylabel('GHI (W/m²)')

plt.subplot(2, 2, 2)
plt.plot(monthly_data.index, monthly_data['DNI'], color='blue')
plt.title('Monthly Average DNI')
plt.xlabel('Month')
plt.ylabel('DNI (W/m²)')

plt.subplot(2, 2, 3)
plt.plot(monthly_data.index, monthly_data['DHI'], color='green')
plt.title('Monthly Average DHI')
plt.xlabel('Month')
plt.ylabel('DHI (W/m²)')

plt.subplot(2, 2, 4)
plt.plot(monthly_data.index, monthly_data['Tamb'], color='red')
plt.title('Monthly Average Tamb')
plt.xlabel('Month')
plt.ylabel('Tamb (°C)')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('sierraleone-bumbuna.csv', parse_dates=['Timestamp'])

# Select relevant columns for correlation analysis
correlation_columns = ['GHI', 'DNI', 'DHI', 'TModA', 'TModB', 'WS', 'WSgust', 'WD']
correlation_data = df[correlation_columns].corr()

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap: Solar Radiation, Temperature, and Wind Conditions')
plt.show()
```

```python
# Pair plot for selected variables
sns.pairplot(df[correlation_columns])
plt.suptitle('Pair Plot: Solar Radiation, Temperature, and Wind Conditions', y=1.02)
plt.show()
```
